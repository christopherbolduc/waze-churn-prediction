# 01 – Foundations of Data Science (PACE Planning)

---

## Project Scenario

As the newest member of the Waze Data Analytics team, I was tasked with outlining a proposal for a churn prediction project.

The request came from May Santner, Data Analysis Manager:

> *"Leadership has asked our team to build a machine learning model that predicts user churn. The model will be based on a sample of Waze user data. An accurate model will help our company prevent user churn, improve user retention, and increase growth."*

This notebook documents my planning process using the PACE (Plan, Analyze, Construct, Execute) framework.

---

## Business Problem

User churn negatively impacts key metrics at Waze — including engagement, advertising revenue, and user growth.  
The goal of this project is to use behavioral data to identify users at risk of churning, and provide actionable insights to improve retention.

---

## Stakeholders

- May Santner — Data Analysis Manager  
- Sylvester Esperanza — Senior Project Manager  
- Chidi Ga — Senior Data Analyst  
- Harriet Hadzic — Director of Data Analysis

---

## Success Criteria

- A plan that clearly outlines how to approach the churn prediction problem
- A proposal aligned with stakeholder goals
- A defined path through the PACE framework

---

## Data Summary

The project uses synthetic Waze user data, with features such as:

- Session frequency
- Trip behavior
- In-app feature use
- Time since last activity
- Churn labels (for supervised ML)

---

## PACE Framework Summary

- **Plan** – Define the business problem, scope, and success criteria  
- **Analyze** – Explore and clean the data, compute descriptive stats  
- **Construct** – Build and evaluate churn prediction models  
- **Execute** – Share findings with stakeholders and recommend actions
